```python
import cv2
import time
import mediapipe as mp
```

```python
cap = cv2.VideoCapture(0)

mpHands = mp.solutions.hands
hands = mpHands.Hands()
mpDraw = mp.solutions.drawing_utils
```

```python
cTime = 0
pTime = 0

while True:
    ret, frame = cap.read()
    frame = cv2.flip(frame, 1)
    
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result = hands.process(img_rgb)
    #print(result.multi_hand_landmarks) 
    # when a hand is detected the values for the landmarks are printed out but None is printed when no hand in detected
    
    # We can then draw the landmarks and the lines connectiong the landmarks for the hand
    if result.multi_hand_landmarks:
        for hand_landmrks in result.multi_hand_landmarks:
            # we then get the Id number and the landmark information for each hand
            for id, lm in enumerate(hand_landmrks.landmark):
                #print(id,lm)
                # the land mark information is normalised, so we converty to pixel vaues
                # this will help us take a specific landmark and perform various tasks
                img_height, img_width, img_channel = frame.shape
                x, y = int(lm.x * img_width), int(lm.y * img_height)
                print(id, x, y)
                # we can use any of these positions eg(id:0, x:435, y:437) to print out any of our landmarks
                if id==4:
                    cv2.circle(frame, (x, y), 15, (255,0,255), cv2.FILLED)
                
            mpDraw.draw_landmarks(frame, hand_landmrks, mpHands.HAND_CONNECTIONS)
            
    
    cTime = time.time()
    fps = 1/(cTime-pTime)
    pTime = cTime
    cv2.putText(frame, f"frames per second{int(fps)}", (10,20), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 1)
    
    
    cv2.imshow("show hands", frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
        
cap.release()
cv2.destroyAllWindows()
```
