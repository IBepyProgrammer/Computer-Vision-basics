```python
import cv2
import mediapipe as mp
import time
```

```python
cap = cv2.VideoCapture("video1.mkv")
pTime = 0

mpFaceDetection = mp.solutions.face_detection
mpDraw = mp.solutions.drawing_utils
faceDetection = mpFaceDetection.FaceDetection()

while True:
    ret, frame = cap.read()
    
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = faceDetection.process(img_rgb)
    print(results)
    
    if results.detections:
        for id, detection in enumerate(results.detections):
#             mpDraw.draw_detection(frame, detection)
#             print(id, detection)
#             print(detection.score)
#             print(detection.loaction_data.relative_bounding_box)             
         
            bboxC = detection.location_data.relative_bounding_box
            img_height, img_width, img_chanels = frame.shape
            bbox = int(bboxC.xmin * img_width), int(bboxC.ymin * img_height), int(bboxC.width * img_width), int(bboxC.height * img_height)
            cv2.rectangle(frame, bbox, (255,0,255), 2)
            cv2.putText(frame, f"Confidence {int(detection.score[0]*100)}", (bbox[0],bbox[1] -20), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
            
    cTime = time.time()
    fps = 1/(cTime-pTime)
    cv2.putText(frame, f"Video_FPS: {int(fps)}", (20,70), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
    cv2.imshow("Detected", frame)
    
    if cv2.waitKey(20) & 0xFF == ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
def decorative_box(frame, bbox, l=20, t=5):
    x, y, w, h = bbox
    x1, y1 = x + w, y + h
    cv2.rectangle(frame, bbox, (255,0,255), 1)
    # for top left x,y
    cv2.line(frame, (x, y), (x +l, y), (0, 255, 0), t)
    cv2.line(frame, (x, y), (x, y +l), (0, 255, 0), t)
    # for top right x1,y
    cv2.line(frame, (x1, y), (x1 -l, y), (0, 255, 0), t)
    cv2.line(frame, (x1, y), (x1, y +l), (0, 255, 0), t)
    # for bottom left x,y1
    cv2.line(frame, (x, y1), (x +l, y1), (0, 255, 0), t)
    cv2.line(frame, (x, y1), (x, y1 -l), (0, 255, 0), t)
     # for bottom right x1,y1
    cv2.line(frame, (x1, y1), (x1 -l, y1), (0, 255, 0), t)
    cv2.line(frame, (x1, y1), (x1, y1 -l), (0, 255, 0), t)
    
    return frame


cap = cv2.VideoCapture(0)
pTime = 0
cTime = 0

mpFaceDetection = mp.solutions.face_detection
mpDraw = mp.solutions.drawing_utils
faceDetection = mpFaceDetection.FaceDetection()

while True:
    ret, frame = cap.read()
    frame = cv2.flip(frame,1)
    
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = faceDetection.process(img_rgb)
    print(results)
    
    if results.detections:
        for id, detection in enumerate(results.detections):
#             mpDraw.draw_detection(frame, detection)
#             print(id, detection)
#             print(detection.score)
#             print(detection.loaction_data.relative_bounding_box)             
         
            bboxC = detection.location_data.relative_bounding_box
            img_height, img_width, img_chanels = frame.shape
            bbox = int(bboxC.xmin * img_width), int(bboxC.ymin * img_height), int(bboxC.width * img_width), int(bboxC.height * img_height)
            #cv2.rectangle(frame, bbox, (255,0,255), 2)
            decorative_box(frame, bbox, l=30, t=10)
            cv2.putText(frame, f"Confidence {int(detection.score[0]*100)}", (bbox[0],bbox[1] -20), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
            
    cTime = time.time()
    fps = 1/(cTime-pTime)
    cv2.putText(frame, f"Video_FPS: {int(fps)}", (20,70), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
    cv2.imshow("Detected", frame)
    
    if cv2.waitKey(20) & 0xFF == ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```
